```python
file_path  = '/Workspace/Users/jjaswanth66@gmail.com/retail_data/Online Retail.csv'
```

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession
    .builder
    .appName('reatail_data')
    .getOrCreate()
)

```

```python
bronze_df = (
    spark.read
    .option('header','true')
    .csv(file_path)
)

display(bronze_df)
```

```python
bronze_df.printSchema()
```

```python
WRITE_MODE_OVERWRITE = "overwrite"

bronze_df.write \
    .format("delta") \
    .mode(WRITE_MODE_OVERWRITE) \
    .saveAsTable('retail_data.bronze.bronze_online_retail')
```
